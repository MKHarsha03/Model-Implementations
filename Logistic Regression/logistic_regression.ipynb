```python
import numpy as np
import pandas as pd
from sklearn import datasets
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
```

```python
class LogisticRegression():
    def __init__(self,lr=0.1):
        self.learning_rate = lr
        self.w = 0
        self.b = 0
    def fit(self,X,y,num_iterations=500,print_cost=False):
        w = np.zeros((X.shape[1],1))
        b = 0
        m = len(y)
        for i in range(num_iterations):
            z = np.dot(X,w) + b
            y_hat = 1/(1+np.exp(-z))

            if(print_cost==True):
                print(f"The cost after {i} iterations is: {(1/m)*np.sum(-y*np.log(y_hat)-(1-y)*np.log(1-y_hat))}")
            
            dw = np.dot(X.T,(y_hat-y))/m
            db = np.sum(y_hat-y)/m

            w = w - (self.learning_rate)*dw
            b = b - (self.learning_rate)*db

        self.w = w
        self.b = b
    
    def predict(self,X,threshold=0.5):
        z = np.dot(X,self.w) + self.b
        predictions = 1/(1+np.exp(-z))
        outputs = (predictions>=threshold)*1.0
        return outputs
        
```

```python
data = datasets.load_breast_cancer()
target = data.target
data = pd.DataFrame(data=data.data,columns=data.feature_names)
data.head()
```

```python
scaler = MinMaxScaler()
X = scaler.fit_transform(data)
X = np.array(X)
y = target.reshape(target.shape[0],1)
```

```python
X_train,X_val,y_train,y_val = train_test_split(X,y,train_size=0.8,test_size=0.2)
model = LogisticRegression()
model.fit(X_train,y_train)
preds = model.predict(X_val)
```
